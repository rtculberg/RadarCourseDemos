```python
import numpy as np
import scipy.io as sio
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable

# Sets the size of all inline plts
plt.rcParams['figure.figsize'] = [15,8]
```

```python
win_path = 'C:/Data/rds_20140426_01_041.mat'

tmp = sio.loadmat(win_path, struct_as_record = False, squeeze_me = True)
data = tmp['seg']
power = 10*np.log10(np.square(np.absolute(data.Data)))
trunc_power = power[0:1000:1,::]
```

```python
fig, ax = plt.subplots()
im = ax.imshow(trunc_power, cmap = plt.cm.Blues, aspect = 'auto', vmin = -180, vmax = -40)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size = "3%", pad = 0.05)
                          
plt.colorbar(im,cax=cax)
```
